Reads raw CSV files and writes to source_to_bronze layer

```python

employee_path   = '/Volumes/workspace/default/upload/Employee-Q1(in).csv'
department_path = '/Volumes/workspace/default/upload/Department-Q1(in).csv'
country_path    = '/Volumes/workspace/default/upload/Country-Q1(in).csv'

print(' Source paths:')
print('  ', employee_path)
print('  ', department_path)
print('  ', country_path)
```

```python

employee_df = spark.read.format('csv') \
    .option('header', 'true') \
    .option('inferSchema', 'true') \
    .load(employee_path)

print('employee_df loaded')
print('Row count:', employee_df.count())
print('Columns  :', employee_df.columns)
display(employee_df)
```

```python

department_df = spark.read.format('csv') \
    .option('header', 'true') \
    .option('inferSchema', 'true') \
    .load(department_path)

print('department_df loaded')
print('Row count:', department_df.count())
print('Columns  :', department_df.columns)
display(department_df)
```

```python

country_df = spark.read.format('csv') \
    .option('header', 'true') \
    .option('inferSchema', 'true') \
    .load(country_path)

print('country_df loaded')
print('Row count:', country_df.count())
print('Columns  :', country_df.columns)
display(country_df)
```

```python

employee_df.write \
    .mode('overwrite') \
    .option('header', 'true') \
    .csv('/Volumes/workspace/default/upload/source_to_bronze/employee')

department_df.write \
    .mode('overwrite') \
    .option('header', 'true') \
    .csv('/Volumes/workspace/default/upload/source_to_bronze/department')

country_df.write \
    .mode('overwrite') \
    .option('header', 'true') \
    .csv('/Volumes/workspace/default/upload/source_to_bronze/country')

print('All 3 files written to source_to_bronze')
```
