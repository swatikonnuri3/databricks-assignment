# utils
Common functions used by all notebooks

```python

from pyspark.sql.functions import *
from pyspark.sql.types import *
import re

print(' Imports done')
```

```python

def camel_to_snake(name):
    s1 = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', name)
    return re.sub('([a-z0-9])([A-Z])', r'\1_\2', s1).lower()


print(camel_to_snake('EmployeeID'))   
print(camel_to_snake('EmployeeName')) 
print(camel_to_snake('DepartmentID')) 
print(camel_to_snake('CountryCode'))  
print(camel_to_snake('DepartmentName')) 
print(camel_to_snake('CountryName'))  
```

```python

def convert_to_snake_case(df):
    new_columns = [camel_to_snake(col) for col in df.columns]
    return df.toDF(*new_columns)

print('convert_to_snake_case defined')
```

```python

def add_load_date(df):
    return df.withColumn('load_date', current_date())

print(' add_load_date defined')
```
